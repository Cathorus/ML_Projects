# Load and Check Data

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("/kaggle/input/spamham/spamham.csv")
```

```python
# replace the null values with a null string
df = df.where((pd.notnull(df)),'')
```

```python
df.shape
```

```python
df.head()
```

```python
df['Category'].value_counts()
```

```python
# 0 = ham , 1 = spam
df.replace({"Category":{"ham":0,"spam":1}},inplace=True)
```

```python
df.head()
```

```python
# separating the data as texts and label
x = df['Message']
y = df['Category']
```

<a id = "2"></a><br>
### train test split

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.2, random_state=42)
```

<a id = "3"></a><br>
### Feature Extraction

```python
# transform the text data to feature vectors that can be used as input to the Logistic regression

feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase='True')

x_train_features = feature_extraction.fit_transform(x_train)
x_test_features = feature_extraction.transform(x_test)

# convert Y_train and Y_test values as integers

y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
print(x_train)
```

```python
print(x_train_features)
```

<a id = "4"></a><br>
# Training

<a id = "5"></a><br>
### Logistic Regression

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train_features,y_train)
```

<a id = "6"></a><br>
### Evaluating the trained model

```python
# prediction on training data
train_data_pred = lr.predict(x_train_features)
train_data_acc = accuracy_score(y_train, train_data_pred)
```

```python
print(f"Accuracy score of training data is : %{train_data_acc*100}")
```

```python
test_data_pred = lr.predict(x_test_features)
test_data_acc = accuracy_score(y_test,test_data_pred)
```

```python
print(f"Accuracy score of test data is : %{test_data_acc*100}")
```

<a id = "7"></a><br>
# Building a Predictive System

```python
input_mail = ["I've been searching for the right words to thank you for this breather. I promise i wont take your help for granted and will fulfil my promise. You have been wonderful and a blessing at all times"]

# convert text to feature vectors
input_data_features = feature_extraction.transform(input_mail)

# making prediction

prediction = lr.predict(input_data_features)
print(prediction)


if (prediction[0]==1):
  print('Ham mail')

else:
  print('Spam mail')
```
