#### 1.Preparing Data

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("/kaggle/input/creditcardfraud/creditcard.csv")
df.shape
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.describe()
```

```python
df['Class'].value_counts() # 0 Normal,1fraudulent Transaction
legit = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
legit.Amount.describe()
```

```python
# compare the values for both transactions
df.groupby('Class').mean()
```

<a id='2'></a>
#### 2.Under-Sampling

```python
legit_sample = legit.sample(n=492)
```

```python
new_df = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_df.head()
```

```python
new_df['Class'].value_counts()
```

```python
x = new_df.drop(columns='Class', axis=1)
y = new_df['Class']
```

<a id='3'></a>
#### 3.Training

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

##### Creating Model


```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

<a id='4'></a>
#### 4.Model Evaluation

##### Accuracy score

```python
x_train_pred = model.predict(x_train)
training_data_acc = accuracy_score(x_train_pred,y_train)
print(f"Accuracy score of training data : {training_data_acc}")
```

```python
x_test_pred = model.predict(x_test)
test_data_acc = accuracy_score(x_test_pred,y_test)
print(f"Accuracy score of training data : {test_data_acc}")
```
